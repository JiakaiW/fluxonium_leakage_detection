```python
import numpy as np
import matplotlib.pyplot as plt



def alpha(t, Delta_rd, epsilon_r, kappa):
    return (-1j*epsilon_r / (1j*Delta_rd + kappa/2)) * (1 - np.exp(-(1j*Delta_rd + kappa/2)*t))

plt.figure(figsize=(8, 6))
tlist = np.linspace(0, 10000, 1000) 

epsilon_r = 0.003 * 2*np.pi


Delta_compu = -0.001 * 2*np.pi
Delta_0 = 0.000 * 2*np.pi

chi = abs(Delta_compu-Delta_0)/2

kappa =  chi/2


alpha_t_values = alpha(tlist, Delta_compu, epsilon_r, kappa)
plt.plot(alpha_t_values.imag, alpha_t_values.real, color='red',label='computational states')

alpha_t_values = alpha(tlist, Delta_0, epsilon_r, kappa)
plt.plot(alpha_t_values.imag, alpha_t_values.real, color='green',label='leakage 0')



plt.title(r'analytical solution of $\dot{\alpha} = -i \Delta_{rd} \alpha - i \epsilon_r - \frac{\kappa}{2} \alpha$')
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.legend()
plt.grid(True)
plt.gca().set_aspect('equal', 'box')
plt.show()
```
