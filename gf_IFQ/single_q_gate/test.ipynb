```python
qbt = gfIFQ(EJ = 3,EC = 3/3, EL = 3/25,truncated_dim=4)
qbt.fluxonium.plot_wavefunction(which=[0,1,2,3,4,6,7,8])

e_ops = []
for ql in range(qbt.truncated_dim):
    e_ops.append(qutip.ket2dm(qutip.basis(qbt.truncated_dim, ql)))
```

```python
t_stop = 100
tlist = np.linspace(0,t_stop, 200)
initial_states = [
    qutip.basis(qbt.truncated_dim, 0),
    qutip.basis(qbt.truncated_dim, 2)
]
results = qbt.run_qutip_mesolve_parrallel(
        initial_states=initial_states,
        tlist=tlist,
        drive_terms = qbt.get_Raman_drive_terms(i=0,
                                                j=3,
                                                k=2,
                                                detuning= 0.03,
                                                amp_scaling_factor=1.0,
                                                amp1_scaling_factor = 1,
                                                amp2_scaling_factor = 1,
                                                t_duration=t_stop,
                                                shape = 'sin^2'),
        c_ops = None,
        e_ops=e_ops
    )

fig, axes = plt.subplots(2,2,figsize = (6,6))
for j in range(2):
    for i in range(qbt.truncated_dim):
        axes[j,0].plot(tlist,results[j].expect[i],label= f"{i}")
        axes[j,1].plot(tlist,results[j].expect[i],label= f"{i}")
    axes[j,1].legend()
    axes[j,0].set_yscale('log')
    axes[j,0].set_ylim(1e-9,2)
```

```python
t_stop = 100
tlist = np.linspace(0,t_stop, 200)
initial_states = [
    qutip.basis(qbt.truncated_dim, 0),
    qutip.basis(qbt.truncated_dim, 2)
]


terms = qbt.get_Raman_drive_terms(i=0,
                            j=3,
                            k=2,
                            detuning= 0,
                            amp_scaling_factor=1,
                            t_duration=t_stop,
                            shape = 'sin^2')
num = 20
amp_array = np.linspace(0.4,1.5,40)
detuning_array = np.linspace(0,0.04,num)
list_of_systems = []
list_of_kwargs = []
for y0 in initial_states[:1]:
    for amp_scaling_factor in amp_array:
        for detuning in detuning_array:
            terms_copy = deepcopy(terms)
            for term in terms_copy:
                term.set_pulse_shape_arg('amp',term.pulse_shape_args_without_id['amp'] * amp_scaling_factor)    
                term.set_pulse_shape_arg('w_d',term.pulse_shape_args_without_id['w_d'] - detuning) 
            list_of_systems.append(qbt)
            list_of_kwargs.append({
                    'y0':y0,
                    'tlist':tlist,
                    'drive_terms': terms_copy,
                    'e_ops': e_ops
                })
```

```python
results = run_parallel_ODEsolve_and_post_process_jobs_with_different_systems(
    list_of_systems = list_of_systems,
    list_of_kwargs = list_of_kwargs,
    max_workers = 7
)
```

```python

pop_list = []
for level in [0,1,2]:
    pop_list.append([])
    for i,amp_scaling_factor in enumerate(amp_array):
        pop_list[-1].append([])
        for j, detuning in enumerate(detuning_array):
            pop_list[-1][-1].append(results[i*20+j].expect[level][-1])

    pop_list[-1] = np.array(pop_list[-1])
```

```python
fig,axes = plt.subplots(1,4,figsize=(16,5),width_ratios=[3,3,3,0.5],)
plt.subplots_adjust(wspace= 0.3)
for level in [0,1,2]:
    im = axes[level].imshow(pop_list[level], cmap= 'viridis',extent=[ amp_array[0], amp_array[-1],  detuning_array[0], detuning_array[-1]], aspect='auto',origin='lower')
    axes[level].set_ylabel('detuning (GHz)')
    axes[level].set_xlabel('amp or Rabi frequency')
    axes[level].set_title(rf'$\langle {level} \rangle$')

# Create a ScalarMappable and use it to create the colorbar
sm = plt.cm.ScalarMappable(cmap='viridis')
sm.set_array([])  # Only needed to avoid an error

cbar = fig.colorbar(sm, cax=axes[-1])
plt.show()
```
