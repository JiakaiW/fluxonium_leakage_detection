```python
import scqubits
import numpy as np
```

```python
# qbt = scqubits.Fluxonium(EJ = 5.65,EC=0.95,EL=0.292,flux=0.5,cutoff=110)

# qbt = scqubits.Fluxonium(EJ = 3,EC=3/4,EL=3/20.5,flux=0,cutoff=110)

qbt = scqubits.Fluxonium(EJ = 3,EC=3/4,EL=3/20.5,flux=0,cutoff=110)

qbt.plot_wavefunction(which = [0,1,2,3,4])
```

```python
qbt.plot_matrixelements(
    'phi_operator',
    show_numbers=True,
    show3d=False
)
```

```python
qbt.plot_matrixelements(
    'n_operator',
    show_numbers=True,
    show3d=False
)
```

```python
qbt.plot_matelem_vs_paramvals(operator='n_operator',param_name='flux', param_vals=np.linspace(0, 1, 80),select_elems=[(1,0),(2,1),(2,0)])
```

```python
qbt.plot_matelem_vs_paramvals(operator='phi_operator',param_name='flux', param_vals=np.linspace(0, 1, 80),select_elems=[(1,0),(2,1),(2,0)])
```

```python
qbt_integer = scqubits.Fluxonium(EJ = 3,EC=3/4,EL=3/20.5,flux=0,cutoff=110)
qbt_conven = scqubits.Fluxonium(EJ = 3,EC=3/4,EL=3/20.5,flux=0.5,cutoff=110)

print(f"upside-down vs conven:")
print(f"upside-down phi op: {qbt_integer.matrixelement_table('phi_operator')[1,2]}")
print(f"conventional phi op: {qbt_conven.matrixelement_table('phi_operator')[0,1]}")
print(f"upside-down n op: {qbt_integer.matrixelement_table('n_operator')[1,2]}")
print(f"conventional n op: {qbt_conven.matrixelement_table('n_operator')[0,1]}")

print("\n\n integer vs conven")
print(f"integer phi op: {qbt_integer.matrixelement_table('phi_operator')[0,1]}")
print(f"conventional phi op: {qbt_conven.matrixelement_table('phi_operator')[0,1]}")
print(f"integer n op: {qbt_integer.matrixelement_table('n_operator')[1,0]}")
print(f"conventional n op: {qbt_conven.matrixelement_table('n_operator')[0,1]}")

```
