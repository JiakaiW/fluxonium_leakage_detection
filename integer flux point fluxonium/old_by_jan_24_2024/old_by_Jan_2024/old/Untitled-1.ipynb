```python
import numpy as np
from IPython.display import clear_output

p = 0.006
size = 10000

def get_T():
    get_chara = False
    counter = 0
    five_star_non_ideal = False
    while get_chara == False:
        sample = np.random.rand() < p
        if sample == 0:
            counter += 1
        else:
            if five_star_non_ideal:
                get_chara = True
            else:
                ideal = np.random.rand() <= 0.5
                if ideal:
                    get_chara = True
                else:
                    five_star_non_ideal = True
                    counter += 1
    return counter

num_realities = 10000
T_list = []
for _ in range(num_realities):
    T_list.append(get_T())

exp_T = sum(T_list) / num_realities

print(f"Expected T: {exp_T}")
```
