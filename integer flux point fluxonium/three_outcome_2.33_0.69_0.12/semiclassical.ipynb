```python
import numpy as np
import matplotlib.pyplot as plt



def alpha(t, Delta_rd, epsilon_r, kappa):
    return (-1j*epsilon_r / (1j*Delta_rd + kappa/2)) * (1 - np.exp(-(1j*Delta_rd + kappa/2)*t))

plt.figure(figsize=(8, 6))
tlist = np.linspace(0, 7000, 1000) 
Delta_rd = 0.001  # Detuning of the readout resonator
epsilon_r = -0.005  # External driving term
kappa = 0      # Decay rate of the resonator
alpha_t_values = alpha(tlist, Delta_rd, epsilon_r, kappa)
plt.plot(alpha_t_values.imag, alpha_t_values.real, label='zero kappa')

Delta_rd = 0.001  # Detuning of the readout resonator
epsilon_r = -0.005  # External driving term
kappa = 4e-5      # Decay rate of the resonator
alpha_t_values = alpha(tlist, Delta_rd, epsilon_r, kappa)
plt.plot(alpha_t_values.imag, alpha_t_values.real, label='small kappa')

Delta_rd = 0.001  # Detuning of the readout resonator
epsilon_r = -0.005  # External driving term
kappa = 1e-4      # Decay rate of the resonator
alpha_t_values = alpha(tlist, Delta_rd, epsilon_r, kappa)
plt.plot(alpha_t_values.imag, alpha_t_values.real, label='big kappa')

Delta_rd = 0.0001  # Detuning of the readout resonator
epsilon_r = -0.005  # External driving term
kappa = 1e-4      # Decay rate of the resonator
alpha_t_values = alpha(tlist, Delta_rd, epsilon_r, kappa)
plt.plot(alpha_t_values.imag, alpha_t_values.real, label='big kappa')

Delta_rd = -0.0005  # Detuning of the readout resonator
epsilon_r = -0.005  # External driving term
kappa = 1e-4      # Decay rate of the resonator
alpha_t_values = alpha(tlist, Delta_rd, epsilon_r, kappa)
plt.plot(alpha_t_values.imag, alpha_t_values.real, label='big kappa')


plt.title(r'analytical solution of $\dot{\alpha} = -i \Delta_{rd} \alpha - i \epsilon_r - \frac{\kappa}{2} \alpha$')
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.legend()
plt.grid(True)
plt.gca().set_aspect('equal', 'box')
plt.show()
```
