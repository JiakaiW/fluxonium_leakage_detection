```python
import numpy as np
import matplotlib.pyplot as plt



def alpha(t, Delta_rd, epsilon_r, kappa):
    return (-1j*epsilon_r / (1j*Delta_rd + kappa/2)) * (1 - np.exp(-(1j*Delta_rd + kappa/2)*t))

plt.figure(figsize=(8, 6))
tlist = np.linspace(0, 7000, 1000) 

epsilon_r = -0.005
Delta_compu = -0.001
Delta_0 = 0.0001
Delta_3 = -0.0003

Delta_rd = Delta_compu  # Detuning of the readout resonator
epsilon_r = epsilon_r  # External driving term
kappa = 0      # Decay rate of the resonator
alpha_t_values = alpha(tlist, Delta_rd, epsilon_r, kappa)
plt.plot(alpha_t_values.imag, alpha_t_values.real, color='red',label='computational states')

Delta_rd = Delta_0  # Detuning of the readout resonator
epsilon_r = epsilon_r  # External driving term
kappa = 1e-4      # Decay rate of the resonator
alpha_t_values = alpha(tlist, Delta_rd, epsilon_r, kappa)
plt.plot(alpha_t_values.imag, alpha_t_values.real, color='green',label='leakage 0')

Delta_rd = Delta_3  # Detuning of the readout resonator
epsilon_r = epsilon_r  # External driving term
kappa = 1e-4      # Decay rate of the resonator
alpha_t_values = alpha(tlist, Delta_rd, epsilon_r, kappa)
plt.plot(alpha_t_values.imag, alpha_t_values.real, color='orange', label='leakage 3')


plt.title(r'analytical solution of $\dot{\alpha} = -i \Delta_{rd} \alpha - i \epsilon_r - \frac{\kappa}{2} \alpha$')
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.legend()
plt.grid(True)
plt.gca().set_aspect('equal', 'box')
plt.show()
```
