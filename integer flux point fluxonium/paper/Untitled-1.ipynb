```python
import uuid
import importlib.util

# Step 1: Generate a unique file name using UUID
file_name = str(uuid.uuid4())
file_path = f"{file_name}.py"

# Step 2: Write your functions to the file
code = """
def func1():
    return "Function 1"

def func2():
    return "Function 2"
"""

with open(file_path, 'w') as file:
    file.write(code)

# Step 3: Import the module dynamically
spec = importlib.util.spec_from_file_location(file_name, file_path)
module = importlib.util.module_from_spec(spec)
spec.loader.exec_module(module)

# Now you can use func1 and func2
result1 = module.func1()
result2 = module.func2()

print(result1, result2)
```
